```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
ps.stem('loved')
```

```python
def stem(text):
  L=[]
  for i in text.split():
    base=ps.stem(i)
    L.append(base)
  return " ".join(L) # converting list into string
```

```python
stem(text)
```

```python
movies_rem['tags']=movies_rem['tags'].apply(stem)
```

## For movies recommandation system tags are converted into vector, for movies
recommandation lesser the distance between two vector the more similar movies to recomand


### To calculate the distance between two vector we use cosin distance for movies recommndation.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector_movies)
```

```python
similarity_movies=cosine_similarity(vector_movies)
```

```python
similarity_movies[0]
```

```python
movies_rem[movies_rem['original_title']=='Avatar'].index[0]
```

```python
sorted(list(enumerate(similarity_movies[0])), reverse=True,key=lambda x : x[1])[1:6]
```

```python
def movies_recommendation(movies):
  movies_index=movies_rem[movies_rem['original_title']==movies].index[0]
  distance=similarity_movies[movies_index]
  movies_list=sorted(list(enumerate(distance)), reverse=True,key=lambda x : x[1])[1:6]

  for i in movies_list:
    index=i[0]
    print(movies_rem.iloc[index].original_title)

```

```python
movies_recommendation('Avatar')
```

```python
movies_recommendation('Avatar')
```
